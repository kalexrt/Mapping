```python
SYSTEM_PROMPT = """
You are a helpful assistant that helps with data mapping tasks. 
You will be given a DataFrame and a reference DataFrame. 
Your task is to fill the 'Ed-Fi Entity' and 'Ed-Fi Attribute' columns in the input DataFrame based on the 'Field' column.
You will use the reference DataFrame to find the corresponding 'Ed-Fi Entity' and 'Ed-Fi Attribute' for each 'Field'.
If a 'Field' does not have a corresponding entry in the reference DataFrame, leave the 'Ed-Fi Entity' and 'Ed-Fi Attribute' columns as NaN.

The output should be a valid JSON and its format is as follows:
{
    "test start": ("StudentAssessment", "AdministrationDate"),
    "test end": (None, None), 
    "student id": ("Student", "StudentUniqueId"),
    "supplemental student id": (None, None), 
}
"""
```

```python
def invoke_llm(query):
    response = runtime.converse(
        modelId="arn:aws:bedrock:us-east-1:654654390449:application-inference-profile/s0naya29upmr",
        messages=[
            {
                "role":"user", 
                "content":[{"text": SYSTEM_PROMPT}]
            },
            {"role": "user", 
             "content": [{"text": query}]
            }
        ]
    )
    return response["output"]["message"]["content"][0]["text"]
```

```python
test = invoke_llm(f"Here is the reference DataFrame:\n{text}\n\n and here is the input DataFrame:\n{input_df.to_string(index=False)}\n\n")
```

```python
print(test)
```

```python
def clean_json_string(input_string):
    """
    Remove all characters before the first '{' in a string.

    Args:
        input_string (str): The input string potentially containing irrelevant characters before JSON.

    Returns:
        str: The cleaned string starting from the first '{' to the last '}'.
    """
    # Find the index of the first '{'
    start_index = input_string.find("{")
    end_index = input_string.rfind("}")

    # Check if both '{' and '}' are found
    if start_index == -1 or end_index == -1:
        raise ValueError("No complete JSON object found in the string")

    return input_string[start_index : end_index + 1]
```

```python
cleaned = clean_json_string(test)
```

```python
print(cleaned)
```

```python
processed = cleaned.replace("'", '"')
```

```python
print(processed)
```

```python
import ast

parsed = ast.literal_eval(processed)
```

```python
parsed
```

```python
rows = []
for field, (entity, attribute) in parsed.items():
    rows.append({
        "Field": field,
        "Ed-Fi Entity": entity,
        "Ed-Fi Attribute": attribute
    })
```

```python
df = pd.DataFrame(rows)
```
